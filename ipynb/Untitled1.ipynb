```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset=pd.read_csv("C:\\Users\\Admin\\Desktop\\CSV\\temperature.csv")
df=pd.read_csv("C:\\Users\\Admin\\Desktop\\CSV\\accuracy\\tempAccuracy.csv")

X = dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=1/3,random_state=101)

from sklearn.linear_model import LinearRegression
LR=LinearRegression()

LR.fit(X_train,y_train)

y_pred_LR=LR.predict(X_test)
diff_LR=y_test-y_pred_LR

res_df1=pd.concat([pd.Series(y_pred_LR),pd.Series(y_test),pd.Series(diff_LR)],axis=1)
res_df1.columns=['Prediction','Original Data','Diff']

print(res_df1)
avgtemp=(sum(res_df1['Prediction']))/len(res_df1['Prediction'])
print(avgtemp)

import sklearn.metrics as sm
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, y_pred_LR), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(y_test, y_pred_LR), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(y_test, y_pred_LR), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(y_test, y_pred_LR), 2)) 
print("R2 score =", round(sm.r2_score(y_test, y_pred_LR), 2))

plt.scatter(df['Name'],df['accuracyRate'])
plt.xlabel('Name')
plt.ylabel("accuracyRate")
plt.show()
#plt.savefig('C:\\Users\\Admin\\Desktop\\CSV\\images\\temp.png')
```
